Преобразуем в необходимый формат.

```python
test["symbol_scores_1"] = test.symbol_scores_1.astype(float)
test["symbol_scores_2"] = test.symbol_scores_2.astype(float)
test["symbol_scores_3"] = test.symbol_scores_3.astype(float)
test["symbol_scores_4"] = test.symbol_scores_4.astype(float)
test["symbol_scores_5"] = test.symbol_scores_5.astype(float)
test["symbol_scores_6"] = test.symbol_scores_6.astype(float)
test["symbol_scores_7"] = test.symbol_scores_7.astype(float)
test["symbol_scores_8"] = test.symbol_scores_8.astype(float)
test["symbol_scores_9"] = test.symbol_scores_9.astype(float)


test["length_scores_1"] = test.length_scores_1.astype(float)
test["length_scores_2"] = test.length_scores_2.astype(float)
test["length_scores_3"] = test.length_scores_3.astype(float)
test["length_scores_4"] = test.length_scores_4.astype(float)
test["length_scores_5"] = test.length_scores_5.astype(float)
test["length_scores_6"] = test.length_scores_6.astype(float)
test["length_scores_7"] = test.length_scores_7.astype(float)
test["length_scores_8"] = test.length_scores_8.astype(float)
test["length_scores_9"] = test.length_scores_9.astype(float)
test["length_scores_10"] = test.length_scores_10.astype(float)
test["length_scores_11"] = test.length_scores_11.astype(float)
```

# Попробуем использовать алгоритмы МО, в качестве таргера будем использовать условия regno = regno_ai.

```python
group_1['Target']=group_1['regno'] == group_1['regno_ai']
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
```

Раскроем параметр symbols.

```python
group_1['symbols'] = group_1.symbols.str[1:-1]
new_F = group_1['symbols'].str.split(',', expand=True)
new_F.columns = ['symbols_1', 'symbols_2', 'symbols_3', 'symbols_4','symbols_5','symbols_6','symbols_7','symbols_8','symbols_9']
Fram = [group_1, new_F]
group_1_1 = pd.concat(Fram, axis=1)
```

```python
group_1_1 = group_1_1.fillna({'symbols_8': 0})
group_1_1 = group_1_1.fillna({'symbols_9': 0})
```

```python
group_1_1
```

Определим параметры, которые будут использованы в модели.

```python
num_cols = ['camera_id', 
             'result_code', 
             'len', 
             'symbol_scores_1', 
             'symbol_scores_2', 
             'symbol_scores_3',
             'symbol_scores_4',
             'symbol_scores_5',
             'symbol_scores_6',
             'symbol_scores_7',
             'symbol_scores_8',
             'symbol_scores_9',
             'length_scores_1',
             'length_scores_2',
             'length_scores_3',
             'length_scores_4',
             'length_scores_5',
             'length_scores_6',
             'length_scores_7',
             'length_scores_8',
             'length_scores_9',
             'length_scores_10',
             'length_scores_11'
             ]

cat_cols = ['regno_recognize',
    'regno_ai',
    'symbols_1',
    'symbols_2',
    'symbols_3',
    'symbols_4',
    'symbols_5',
    'symbols_6',
    'symbols_7',
    'symbols_8',
    'symbols_9'
]

feature_cols = num_cols + cat_cols
target_col = 'Target'
```

```python
X = group_1_1[num_cols + cat_cols].copy()
y = group_1_1[target_col].copy()
```

Делим на трайн и валидацию.

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.7, random_state=42)
```

```python
clf = CatBoostClassifier(random_seed=42, silent=True)
```

Обучаем модель.

```python
clf.fit(X_train, y_train, cat_features=cat_cols)
```

```python
clf_pred = clf.predict_proba(X_val)[:, 1]
```

```python
roc_auc_score(y_val, clf_pred)
```

Все отлично. Теперь обучим модель на всех данных.
